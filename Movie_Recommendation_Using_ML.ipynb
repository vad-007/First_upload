```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['tags']=df['genre']+df['overview']
```

```python
df.head()
```

```python
new_df=df[['id','title','tags']]
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vc=cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
```

```python
vc
```

```python
vc.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim=cosine_similarity(vc)
```

```python
sim
```

```python
new_df[new_df['title']=='The Godfather'].index[0]
```

```python
dist=sorted(list(enumerate(sim[2])),reverse=True,key=lambda x:x[1])
```

```python
dist
```

```python
for i in dist[1:6]:
    print(new_df.iloc[i[0]].title)
```

```python
def recommend(movies):
  index=new_df[new_df['title']==movies].index[0]
  dist=sorted(list(enumerate(sim[index])),reverse=True,key=lambda x:x[1])
  for i in dist[1:6]:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Dilwale Dulhania Le Jayenge')
```
